```python
#to check the distribution of how long ppl work
mask = app_train["DAYS_EMPLOYED"] != 365243
app_train[mask]["DAYS_EMPLOYED"].plot(kind = "kde")
```

```python
#scatter plot b/t income & credit
data = app_train[["AMT_INCOME_TOTAL", "AMT_CREDIT"]]
trans_data = np.log(data).diff().dropna()
plt.scatter(trans_data["AMT_INCOME_TOTAL"], trans_data["AMT_CREDIT"])
plt.title("Changes in log Income vs. log Credit")
```

```python
print (app_train[["AMT_INCOME_TOTAL", "AMT_CREDIT"]].mean())
```

```python
app_train["Credit/Income ratio"] = app_train["AMT_CREDIT"]/app_train["AMT_INCOME_TOTAL"]
data_plt = app_train["Credit/Income ratio"]
x_bin = range(1000)
plt.plot(x_bin,data_plt[0:1000])
```
